```python
import os
import datetime
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import yahoo.src.option_strategy as OptS
```

## Example: SPX

### Straddle

```python
one = OptS.Straddle(ticker="SPX", trade_date="2009-12-11", quantity=-7, target_dtox = 30)
one.build()
fig_h1, ax1 = one.plot_theo_pnl(show=True)
```

### Strangle

```python
two = OptS.Strangle(ticker="SPX", trade_date="2009-12-11", quantity=1, target_dtox = 30)
two.set_strikes(strikes=[1100,1200])
two.build()
fig_h2, ax2 = two.plot_theo_pnl(show=True)
```
